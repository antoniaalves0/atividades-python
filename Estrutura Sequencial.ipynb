#####  Calcule o salário bruto (horas * salario por hora)

```python
salario_h = float(input('quanto vc ganha por hora?'))
horas = int(input('qual numero de horas trabalhadas no mês?'))
salario_bruto = horas * salario_h
print('o seu salario bruto é: ' + str(salario_bruto))
```

##### Calcule o desconto do IR (11% do salário bruto)

```python
salario_h = float(input('quanto vc ganha por hora?'))
horas = int(input('qual numero de horas trabalhadas no mês?'))
salario_bruto = (horas * salario_h) 
ir = salario_bruto * 0.11
print('o seu desconto do IR é de: R$ ' + str(ir) )
```

##### Calcule o desconto do INSS (8% do salário bruto)

```python
salario_h = float(input('quanto vc ganha por hora?'))
horas = int(input('qual numero de horas trabalhadas no mês?'))
salario_bruto = (horas * salario_h) 
inss = salario_bruto * 0.8
print('o seu desconto do INSS é de: R$ ' + str(inss) )
```

##### Calcule o desconto do sindicato (5% do salário bruto)

```python
salario_h = float(input('quanto vc ganha por hora?'))
horas = int(input('qual numero de horas trabalhadas no mês?'))
salario_bruto = (horas * salario_h) 
sindicato = salario_bruto * 0.5
print('o seu desconto do sindicato é de: R$ ' + str(sindicato) )
```

##### Calcule o salário líquido (salário bruto - descontos)

```python
salario_h = float(input('quanto vc ganha por hora?'))
horas = int(input('qual numero de horas trabalhadas no mês?'))
salario_bruto = (horas * salario_h) 
ir = salario_bruto * 0.11
inss = salario_bruto * 0.8
sindicato = salario_bruto * 0.5
descontos = ir + inss + sindicato 
salario_liquido = salario_bruto - descontos
print('o seu salario liquido é de: R$ ' + str(descontos) )
```

#### 14. Faça um programa que peça o tamanho de um arquivo para download (em MB) e a velocidade de um link de Internet (em Mbps), calcule e informe o tempo aproximado de download do arquivo usando este link (em minutos).

Detalhe: MB significa megabyte, Mb (com b minúsculo) significa megabit. Um megabit é 1/8 de um megabyte. 

```python
tamanho = int(input('qual o tamanho do arquivo? em MB'))
veloc = int(input('qual a velocidade do link? em Mbps'))
tempo = tamanho / (veloc/8)
minutos = tempo / 60
print('o tempo aproximado é de ' , minutos , 'minutos')

```
